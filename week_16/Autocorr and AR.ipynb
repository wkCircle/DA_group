```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import statsmodels.tsa as tsa
```

```python
# The plot of white noise

mean = 0
std = 1 
num_samples = 300
samples = np.random.normal(mean, std, size=num_samples)

plt.plot(samples)
plt.show()
```

```python
# The plot of moving average

ma_process = (samples[:-1] + samples[1:])/2

plt.plot(ma_process)
plt.show()
```

```python
# The plot of random walk

rw = [ samples[:t+1].sum() for t in range(len(samples))]

plt.plot(rw)
plt.show()
```
